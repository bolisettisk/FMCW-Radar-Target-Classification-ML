```python
import sys
import os
import joblib
import numpy as np
import random
from datetime import datetime
import matplotlib.pyplot as plt

from tensorflow.keras.utils import to_categorical
from scikeras.wrappers import KerasClassifier
from sklearn.metrics import accuracy_score

from IPython.display import clear_output
```

```python
# current_dir = sys.path.copy()
# library_dir = "..\Data Simulation"
# sys.path.append(os.path.abspath(library_dir))
from FMCW_Data_Simulation_Library import Tx_gen, Get_Target_Params, Rx_mpt_gen, Rx_add_noise, FFT_gen, Target_Data_gen, ML_Model_Tester

# sys.path = current_dir
```

```python
range_max = 200 # Maximum range of the target
c = 3e8 # Speed of light
range_resolution = 1 # FMCW radar range resolution
Nd = 64 # Number of chirps
Nr = 256 # Numnber samples per chirp
fc = 77e9 # FMCW radar carrier frequency
endle_time = 6.3e-6 # Time delay between ramps
range_window = np.hamming(Nr) # hamming window for range FFT data
doppler_window = np.hamming(Nd) # hamming window for doppler FFT data
Shapes_dict = {"Square":0, "Triangle":1} # Required target shapes to simulate
```

```python
target_directory = "ML_Model_Doppler_3dB_Sim"
ML_models_lsit = os.listdir(target_directory)

# ML_models_lsit = ['GBC_Classifier.joblib']
ML_model_names = [i.split(".")[0] for i in ML_models_lsit]

model_accuracies = {}   
for model in ML_model_names:
    model_accuracies[model] = []
model_accuracies


for SNR_dB in np.arange(-15, 5, 0.25):
    if SNR_dB < 0:
        filename = f"m{abs(float(SNR_dB))}dB" 
    else:
        filename = f"{abs(float(SNR_dB))}dB"

    evaluation_data_path = '../data/evaluation_data'

    evaluation_data_file_path = evaluation_data_path + '/' + 'evaluation_data' + '_' + filename + '.npy'
    evaluation_data = np.load(evaluation_data_file_path)
    evaluation_data_reshape = evaluation_data.reshape(evaluation_data.shape[0], -1)

    evaluation_class_file_path = evaluation_data_path + '/' + 'evaluation_class' + '_' + filename + '.npy'
    evaluation_class = np.load(evaluation_class_file_path)
    evaluation_class_cat = to_categorical(evaluation_class, len(Shapes_dict)) # Converting evaluation_class to categorical
    
    for model in ML_model_names:
        print("--- {}dB SNR: {} model! ---".format(SNR_dB, model)) # Printing status
        ML_path = target_directory + '/' + model + '.joblib'
        ML_model = joblib.load(ML_path) # Importing the previously trained machine learning model from the target directory
        
        if model in ['LSTM_Classifier', 'RNN_Classifier']:
            eval_class = evaluation_class_cat
#             print("{}: converted to categorical".format(model))
        else:
            eval_class = evaluation_class
#             print("{}: not converted to categorical".format(model))
        
        if model in ['GBC_Classifier']:
            eval_data = evaluation_data_reshape
            y_pred = ML_model.predict(eval_data)
            accuracy = accuracy_score(y_pred, eval_class)
        else:
            eval_data = evaluation_data
            loss, accuracy = ML_model.evaluate(eval_data, eval_class, verbose=0)
    
        model_accuracies[model].append(accuracy)
    clear_output()
```

```python
x_axis_data = np.arange(-15, 5, 0.25)

plt.figure(figsize=(10, 6))
for i in model_accuracies:
    plt.plot(x_axis_data, model_accuracies[i], label=i)
plt.legend()
plt.xlabel('SNR (dB)')
plt.ylabel('Classification Accuracy')
plt.title('Comparison of Classification Accuracies')
```
